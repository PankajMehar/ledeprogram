# Scraping columbia journalism faculty

```python
url = "http://www.journalism.columbia.edu/page/10?category_ids%5B%5D=2&category_ids%5B%5D=3&category_ids%5B%5D=37"
```

```python
from bs4 import BeautifulSoup
import urllib
```

```python
html_str = urllib.urlopen(url).read()
```

```python
print html_str[:100]
```

```python
document = BeautifulSoup(html_str)
```

```python
li_tags = document.find_all("li")
```

```python
for tag in li_tags:
    print tag
```

```python
ul_tag = document.find("ul", attrs={"class": "experts-list"})
```

```python
li_tags = ul_tag.find_all("li")
for tag in li_tags:
    print tag
```

```python
all_faculty = []
for tag in li_tags:
  # print "here's a tag:", tag
   h4_tag = tag.find('h4')
   if h4_tag is not None:
       a_tag = h4_tag.find('a')
       faculty_name = a_tag.string
       p_tag = tag.find('p', attrs={'class': 'description'})
       if p_tag is not None:
           faculty_title = p_tag.string
           faculty_member = {'name': faculty_name, 'title': faculty_title}
           all_faculty.append(faculty_member)
all_faculty
```
